```python
# import cv
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# read image
img = cv.imread('IMG_6693.jpg')
```

```python
# show the image
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai")
```

```python
# Make the image gray 
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# Show the gray image 
plt.imshow(cv.cvtColor(gray_img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai_gray")
```

```python
# Change the size 
resize_img = cv.resize(img, dsize=(200,200))
```

```python
# Show the image after change
plt.imshow(cv.cvtColor(resize_img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai_resize")
```

```python
print('before',img.shape)
print('after',resize_img.shape)
```

```python
# Draw a circle or rectangle
# Coordinate 
x,y,w,h = 100, 100, 100, 100
# Draw a rectangle
cv.rectangle(img, (x,y,x+w,y+h), color=(0,0,255),thickness=1)
# Draw a circle
cv.circle(img, center=(x+w,y+h),radius=50, color=(255,0,0), thickness=2)
# Show
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai")
```
