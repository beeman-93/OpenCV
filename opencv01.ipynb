```python
# import cv
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# read image
# I put the 'IMG_6693.jpg', and this ipynb file under the same folder. 
# Otherwise, please state the right path. 

img = cv.imread('IMG_6693.jpg')
```

```python
# show the image
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai")
```

```python
# Make the image gray 
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# Show the gray image 
plt.imshow(cv.cvtColor(gray_img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai_gray")
```

```python
# Change the size 
resize_img = cv.resize(img, dsize=(200,200))
```

```python
# Show the image after change
plt.imshow(cv.cvtColor(resize_img, cv.COLOR_BGR2RGB))
plt.title("Avatar_Cai_resize")
```
