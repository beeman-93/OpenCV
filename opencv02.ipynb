```python
# import cv
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Detect a face

def face_detect_test():
    gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)   
    face_detect=cv.CascadeClassifier('haarcascade_frontalface_alt2.xml')
    face = face_detect.detectMultiScale(gray_img)
    for x,y,w,h in face:
        cv.rectangle(img,(x,y),(x+w, y+h), color=(0,0,255), thickness=2)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title("Result")

# I put the 'haarcascade_frontalface_alt2.xml', 'IMG_6693.jpg', and this ipynb file under the same folder. 
# Otherwise, please state the right path. 

img = cv.imread('IMG_6693.jpg')
face_detect_test()
```
